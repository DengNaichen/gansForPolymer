```python
import json
```

```python
with open('3_layers/data.json', 'r') as fp:
    data_3 = json.load(fp)
with open('4_layers/data.json', 'r') as fp:
    data_4 = json.load(fp)
with open('5_half/data.json', 'r') as fp:
    data_5_half = json.load(fp)
with open('yousef_5_lr_0_0001_dup/data.json', 'r') as fp:
    data_5 = json.load(fp)
with open('5_layers_double/data.json', 'r') as fp:
    data_5_double = json.load(fp)
with open('6_layers/data.json', 'r') as fp:
    data_6 = json.load(fp)
a = [data_3, data_4, data_5_half, data_5, data_5_double, data_6]
last_key = list(data_6.keys())[-1]
c = data_6.copy()[last_key]
b = {}
for i, _ in c.items():
    b[f'{i}'] = []
```

```python
for i in a:
    last_key = list(i.keys())[-1]
    i = i[last_key]
    for m, n in b.items():
        b[f'{m}'].append(i[f'{m}'])
```

```python
print(b)
```

```python
import matplotlib.pyplot as plt
import numpy as np

for i, j in b.items():
    if str(i) == "count_0" or str(i) == "count_0.5" or str(i) == "count_1":
        pass
    else:
        if str(i) == 'rg2':
            plt.plot(np.array(b[f'{i}'])/6.7946684375)
            plt.plot([1,1,1,1,1,1])
        elif str(i) == 'rg4':
            plt.plot(np.array(b[f'{i}'])/82.47559951171876)
        elif str(i) == 'rg6':
            plt.plot(np.array(b[f'{i}'])/1368.04750044487)
        elif str(i) == 'unique' or str(i) == 'overlap':
            plt.plot(np.array(b[f'{i}'])/16000)
            plt.ylim([0, 1])
        else:
            plt.plot(np.array(b[f'{i}'])/16000)
        plt.title(f'{i}')
        plt.xticks([0, 1, 2, 3, 4, 5],
                   [r'three', r'four', r'5 half', r'five', r'five double(3700)', r'six'])
        plt.show()
        plt.clf()
```
