```python
import matplotlib.pyplot as plt
import json
```

```python
with open('5_half_lr_0_001/gen_loss.json', 'r') as fp:
    gen_loss = json.load(fp)
with open('5_half_lr_0_001/disc_loss.json', 'r') as fp:
    disc_loss = json.load(fp)
```

```python
gen_loss
```

```python
import numpy as np
gen_avg = []
disc_avg = []
for i in gen_loss:
    gen_avg.append(np.mean(np.array(gen_loss[i])))
    disc_avg.append(np.mean(np.array(disc_loss[i])))
```
